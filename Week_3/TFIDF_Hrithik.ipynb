```python
import pandas as pd
```

```python
# Importing data frame
df = pd.read_csv("CarTalkCommunityPostProcessed.csv")
df.head()
```

```python
data = []
for review in df['cooked']:
    if isinstance(review, float):
        continue
    data.append(review.split(" "))
```

```python
def computeTFReview(review):
    reviewTFDict = {}
    for word in review:
        if word in reviewTFDict:
            reviewTFDict[word] += 1
        else:
            reviewTFDict[word] = 1
    for word in reviewTFDict:
        reviewTFDict[word] = reviewTFDict[word] / len(review)
    return reviewTFDict
```

```python
tfDict = {}
reviewNo = 1
for review in data:
    tfDict['review: ' + str(reviewNo)] = computeTFReview(review)
    reviewNo += 1
```

```python
tfDict['review: 3']
```

```python
def computeCountDict(tfDict):
    countDict = {}
    for review in tfDict:
        for word in tfDict[review]:
            if word in countDict:
                countDict[word] += 1
            else:
                countDict[word] = 1
    return countDict
```

```python
countDict = computeCountDict(tfDict)
```

```python
countDict['head']
```

```python
def computeIDF(countDict, n):
    import math
    idfDict = {}
    for word in countDict:
        idfDict[word] = math.log(n / countDict[word])
    return idfDict    
```

```python
idfDict = computeIDF(countDict, len(data))
idfDict
```

```python
def computeTFIDF(review, idfDict):
    tfidfDict = {}
    for word in review:
        tfidfDict[word] = review[word] * idfDict[word]
    return tfidfDict
    
```

```python
# tfidfDict = [computeTFIDF(tfDict[review], idfDict) for review in tfDict]
tfidfDict = {}
review_no = 1
for review in tfDict:
    tfidfDict['review: ' + str(review_no)] = computeTFIDF(tfDict[review], idfDict)
    review_no += 1
```
